```python
# Dependencies and Setup
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data-checkpoint.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data-checkpoint.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_data = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
merged_data = pd.merge(clinical_trial_data, mouse_drug_data, how = "left", on = "Mouse ID")

# Display the data table for preview
merged_data.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
merged_data_mean = merged_data.groupby(["Drug", "Timepoint"]).mean()["Tumor Volume (mm3)"]
merged_data_mean
# Convert to DataFrame
merged_data_mean_df = pd.DataFrame(merged_data_mean)
merged_data_mean_df = merged_data_mean_df.reset_index()
# Preview DataFrame
merged_data_mean_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
merged_data_stderror = merged_data.groupby(["Drug", "Timepoint"]).sem()["Tumor Volume (mm3)"]

# Convert to DataFrame
merged_data_stderror_df = pd.DataFrame(merged_data_stderror)
merged_data_stderror_df = merged_data_stderror_df.reset_index()

# Preview DataFrame
merged_data_stderror_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames

merged_data_mean_df_pivot  = merged_data_mean_df.pivot(index = "Timepoint", columns = "Drug")["Tumor Volume (mm3)"]
merged_data_mean_df = merged_data_mean_df.reset_index()

merged_data_stderror_df = merged_data_stderror_df.reset_index()
merged_data_stderror_df_pivot  = merged_data_stderror_df.pivot(index = "Timepoint", columns = "Drug")["Tumor Volume (mm3)"]
# Preview that Reformatting worked
merged_data_mean_df_pivot.head()
```

```python
merged_data_stderror_df_pivot.head()
```

```python
# Generate the Plot (with Error Bars)
# Generate the Plot (with Error Bars)
plt.errorbar(merged_data_mean_df_pivot.index, merged_data_mean_df_pivot["Capomulin"], yerr=merged_data_stderror_df_pivot["Capomulin"], color="r", marker="o", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(merged_data_mean_df_pivot.index, merged_data_mean_df_pivot["Infubinol"], yerr=merged_data_stderror_df_pivot["Infubinol"], color="b", marker="^", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(merged_data_mean_df_pivot.index, merged_data_mean_df_pivot["Ketapril"], yerr=merged_data_stderror_df_pivot["Ketapril"], color="g", marker="s", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(merged_data_mean_df_pivot.index, merged_data_mean_df_pivot["Placebo"], yerr=merged_data_stderror_df_pivot["Placebo"], color="k", marker="d", markersize=5, linestyle="dashed", linewidth=0.50)

plt.title("Metastatic Spread During Treatment")
plt.ylabel("Met. Sites")
plt.xlabel("Treatment Duration (Days)")
plt.grid(axis='y')
plt.legend(['Capomulin', 'Infubinol', 'Ketapril', 'Placebo'], loc="best", fontsize="small", fancybox=True)

# Save the Figure
plt.savefig("Fig2.png")

# Show the Figure
plt.show()


```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
survival_rates = merged_data.groupby(["Drug","Timepoint"]).count()["Tumor Volume (mm3)"]

# Convert to DataFrame
survival_rates_df = pd.DataFrame({"Mouse Count":survival_rates})

# Preview DataFrame
survival_rates_df.head(10).reset_index()
```

```python
# Minor Data Munging to Re-Format the Data Frames
survival_rates_df = survival_rates_df.reset_index()
survival_rates_df_pivot  = survival_rates_df.pivot(index = "Timepoint", columns = "Drug")["Mouse Count"]
# Preview the Data Frame
survival_rates_df_pivot.head()
```

```python

# Generate the Plot (Accounting for percentages)
plt.plot(100 * survival_rates_df_pivot["Capomulin"] / 25, "ro", linestyle="dashed", markersize=5, linewidth=0.50)
plt.plot(100 * survival_rates_df_pivot["Infubinol"] / 25, "b^", linestyle="dashed", markersize=5, linewidth=0.50)
plt.plot(100 * survival_rates_df_pivot["Ketapril"] / 25, "gs", linestyle="dashed", markersize=5, linewidth=0.50)
plt.plot(100 * survival_rates_df_pivot["Placebo"] / 25 , "kd", linestyle="dashed", markersize=6, linewidth=0.50)
plt.title("Survival During Treatment")
plt.ylabel("Survival Rate (%)")
plt.xlabel("Time (Days)")
plt.grid(True)
plt.legend(['Capomulin', 'Infubinol', 'Ketapril', 'Placebo'], loc="best", fontsize="small", fancybox=True)

# # Save the Figure
# plt.savefig("analysis/Fig3.png")

# Show the Figure
plt.show()

```
